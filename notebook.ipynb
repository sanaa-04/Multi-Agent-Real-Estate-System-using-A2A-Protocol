# Federated Multi-Agent Real Estate System (A2A Protocol)

This notebook demonstrates a federated multi-agent system built with **LangGraph**, **A2A Protocol**, and **Ollama (gemma2)**.

## Architecture
1. **Concierge Agent**: The orchestrator that discovers agents and manages the workflow.
2. **Customer Agent**: Specialized in onboarding customers.
3. **Deal Agent**: Specialized in property data collection.
4. **Marketing Agent**: Uses RAG to generate and store market insights.

## Communication: A2A Protocol
Each agent publishes an `agent.json` (Agent Card) and follows a standardized task-based request/response pattern.
